```python
import numpy as np
import random as r
import math
```

```python

def creaScacchiera(dimensione):
    scacchiera = np.full((dimensione,dimensione), " ", dtype=str)

    for i in range(dimensione):
        scacchiera[i,i]  = 'Q'
    return scacchiera
```

```python
def tweak( scacchiera):
    s = scacchiera.campo.copy()
    while 1 :
        x = r.randint(0 ,scacchiera.dimensione -1)
        y = r.randint(0,scacchiera.dimensione-1)

        if ( y != x ):
            break

    temp = s[:,x].copy()
    s[:,x] = s[:,y]
    s[:,y] = temp
    scacchiera.setCampo(s)

    return Scacchiera( scacchiera.dimensione , s)
```

```python
def calcoloConflitti(scacchiera):
    s = scacchiera.campo
    d = scacchiera.dimensione

    conflitti = 0
    for i in range(d):
        for j in range(d):
            
             if( s[i,j] == 'Q'):
                conflitti += calcolo(s,i,j,d-1)


    return conflitti

            
def calcolo( s, riga , colonna ,d ):

    moves = [(1,1) , (1,-1) , (-1,1), (-1 ,-1)]
    conflittiParziali = 0 
    for move in moves:
    
        i= riga + move[0]
    
        j= colonna + move[1]
      #  print(i,j)
        while( not (i<0 or i>d or j<0 or j>d)):
            if(s[i,j] == 'Q'):
                conflittiParziali += 1
        #        print(f"Trovato q nella posizione ({i},{j}) , a partire dalla riga e colonna {riga}{colonna}")
            i += move[0]
            j += move[1]

    
    return conflittiParziali




```

```python
class Scacchiera():

    def __init__(self, dimensione, campo=None):
        self.dimensione = dimensione
        if (campo is not None): self.campo = campo
        else: 
            self.campo = np.full((dimensione,dimensione), "0", dtype=str)
            self.inizializza(self.campo)

    def inizializza(self, scacchiera):
        for i in range(self.dimensione):
            scacchiera[i,i]  = 'Q'


    def setCampo( self , campo):
        self.campo = campo
            
    def cleanScacchiera(self):
        self.campo = np.full((self.dimensione,self.dimensione), " ", dtype=str)

    def setMove(self, riga , colonna):
        self.campo[riga,colonna] = 'Q'

    
    def stampa(self):
        print(self.campo)

    def evaluation(self):
        return 1/(calcoloConflitti(self)+1)
    
    def tweak(self):
        return tweak(self)

            
```

```python
s = Scacchiera(4)

s.cleanScacchiera()

s.setMove(0,1)
s.setMove(1,3)
s.setMove(2,0)
s.setMove(3,2)

s.stampa()
#0 1 
#1 4 
#2 0
#3 2
g = s.tweak()

g.stampa()
```

```python
def simulatedAnnealing( scacchiera , T , numeroIterazioni):

    t = 0
    currentState =  scacchiera
    bestState = Scacchiera(currentState.dimensione , currentState.campo)
    while ( t< numeroIterazioni  and currentState.evaluation()!=1.0):

        nextState = currentState.tweak()

        
        if ( nextState.evaluation() >= currentState.evaluation()):
            currentState = Scacchiera( nextState.dimensione, nextState.campo)
            if ( currentState.evaluation() > bestState.evaluation()):
                bestState = Scacchiera(currentState.dimensione , currentState.campo)
                print(f"Il best state è \n{bestState.campo}\ncon valutazione{bestState.evaluation()}")

        else:
            delta = currentState.evaluation() - nextState.evaluation()
            expValue = delta/T
            p = math.exp(-expValue)
            if ( r.random() < p):
                currentState = Scacchiera(nextState.dimensione , nextState.campo)


        t+=1    
        T = T*0.99
        print(f"-----------passo {t}----------------")

            

```

```python
s = Scacchiera(10)

simulatedAnnealing(s,10000,100000)


```
